```python
import re

def parse_data_with_numbers(data):
    lines = data.strip().split('\n')
    state = lines[0].split(' ')[-1]

    # The header is assumed to be in the second line, containing "LGA" and "Chairman"
    header = lines[1].split('-')
    if len(header) != 2:
        raise ValueError("Invalid header format")

    header1 = header[0].strip()
    header2 = header[1].strip()

    result = []

    for line in lines[2:]:
        values = line.split(' - ')
        if len(values) != 2:
            continue  # Skip lines without the expected format

        number_lga = values[0].strip()
        chairman = values[1].strip()

        # Split the number and LGA using a dot
        number, lga = number_lga.split(".", 1)

        item = {
            "state": state,
            header1.lower(): lga,
            header2.lower(): chairman,
        }
        result.append(item)

    import json
    json_result = json.dumps(result, indent=2)
    print(json_result)

def convert_string_to_json(data):
    lines = data.strip().split('\n')

    result = []

    state = lines[0].split(' ')[-1]

    header = [header.strip() for header in lines[1].split('-')]
    header1, header2 = header[0], header[1]

    for line in lines[2:]:
        values = line.split('-')
        lga, chairman = values[0].strip(), values[1].strip()
        item = {
            "state": state,
            header1.lower(): lga,
            header2.lower(): chairman,
        }
        result.append(item)

    import json
    json_result = json.dumps(result, indent=2)
    print(json_result)
```
